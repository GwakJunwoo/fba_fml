Margin

1) $\overrightarrow{x_{p}}=\vec{x}-\vec{d}$

2) $\vec{d}=\alpha \cdot \vec{w}$

$$
\begin{aligned}
&w^{\top} x_{p}+b_{0}=w^{\top}(\vec{x}-\vec{d})+b_{0}=w^{\top}(\vec{x}-\alpha \cdot \vec{w})+b_{0}=0 \\
&\Rightarrow \alpha=\frac{w^{\top} x+b_{0}}{w^{\top} w} \\
&\Rightarrow d=\frac{w^{\top} x+b_{0}}{w^{\top} w} \cdot \vec{w} \\
&\|d\|=\sqrt{d^{\top} d}=\sqrt{\alpha^{2} \cdot w^{\top} w}=\alpha \cdot \sqrt{w^{\top} w} \\
&=\frac{w^{\top} x_{n}+b_{0}}{\|w\|}
\end{aligned}
$$
